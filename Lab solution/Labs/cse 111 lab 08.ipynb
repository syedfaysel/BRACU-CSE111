```python
# Task 01

class RealNumber:
    
    def __init__(self, r=0):
        self.__realValue = r  
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())    
    
# write your code here
class ComplexNumber(RealNumber):
    def __init__(self, r = 1, i = 1):
        super().__init__(r)
        self.imaginary = float(i)

    def __str__(self):
        msg = f"RealPart:{float(self.getRealValue())}\nImaginary Part: {self.imaginary}"
        return msg
        
# code ends here

cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
# Task 2


class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)

# Write your code here

class ComplexNumber(RealNumber):
    def __init__(self, data, i):
        if (type(data) == RealNumber):
            data = data.number 
        super().__init__(data)
        self.imaginary = i


    def __add__(self, cmplxNum2):
        # return ComplexNumber((self.number + cmplxNum2.number) , (self.imaginary + cmplxNum2.imaginary))
        r = self.number + cmplxNum2.number
        i = self.imaginary + cmplxNum2.imaginary
        return ComplexNumber(r, i)
    
    def __sub__(self, cmplxNum2):
        return ComplexNumber((self.number - cmplxNum2.number) , (self.imaginary - cmplxNum2.imaginary))


    def __str__(self):
        if (self.imaginary < 0):
            return f"{self.number} - {abs(self.imaginary)}i"
        else:
            return f"{self.number} + {abs(self.imaginary)}i" 

# code ends here

r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
# Task 5

class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
        
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

# write your code here
class ScienceExam(Exam):
    def __init__(self, marks, time, *parts):
        super().__init__(marks)
        self.time = time
        self.sub = parts

    def examSyllabus(self):
        s = f"{super().examSyllabus()}"
        for i in self.sub:
            s += f", {i}"
        return s

    def examParts(self):
        s = f"{super().examParts()}"
        for i in range(len(self.sub)):
            s += f"Part {i+3} - {self.sub[i]}\n"
        return s


    def __str__(self):
        s = f"Marks: {self.marks} Time: {self.time} minutes "
        s += f"Number of parts: {len(self.sub)+2}"
        return s

# code ends here
    
    
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================') 
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
# Task 6


class Shape3D:

  pi = 3.14159
  def __init__(self, name = 'Default', radius = 0):
    self._area = 0
    self._name = name
    self._height = 'No need'
    self._radius = radius   

  def calc_surface_area(self):
    return 2 * Shape3D.pi * self._radius

  def __str__(self):
      return "Radius: "+str(self._radius)

# write your code here

class Sphere(Shape3D):

    def __init__(self, name = "Default", radius = 0):
        super().__init__(name, radius)
        self._name = "Sphere"
        print(f"Shape name: {self._name}, Area Formula: 4 * pi * r * r")

    def calc_surface_area(self):
        self._area = 4 * Shape3D.pi * self._radius * self._radius
        return self._area
    def __str__(self):
        return f"Radius: {self._radius}, Height: {self._height}\nArea: {self._area}"

class Cylinder(Shape3D):

    def __init__(self, name = "Default", radius = 0, height = 0):
        super().__init__(name, radius)
        self._name = "Cylinder"
        self._height = height
        print(f"Shape name: {self._name}, Area Formula: 4 * pi * r * (r + h)")
    
    def calc_surface_area(self):
        self._area = 2 * Shape3D.pi * self._radius * (self._radius + self._height)
        return self._area

    def __str__(self):
        return f"Radius: {self._radius}, Height: {self._height}\nArea: {self._area}"

# code ends here

sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
# Task 7

class PokemonBasic:

  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type

  def get_type(self):
    return 'Main type: ' + self.type

  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'

  def __str__(self):
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness


# Write your code here
class PokemonExtra(PokemonBasic):

    def __init__(self, name, hp, w, t, *args):
        super().__init__(name, hp, w, t)
        self.secType = []
        self.otherMoves = ''
        for i in args:
            if type(i) == tuple:
                self.otherMoves = i
            else:
                self.secType.append(i)

    def get_type(self):
        s = super().get_type()
        if len(self.secType) != 0:
            s += f", Secondary type: "
            for i in self.secType:
                s += i + ","
        return s

    def get_move(self):
        s = super().get_move()
        
        if len(self.otherMoves) != 0:
            s += "\nOther Move: "
            for i in self.otherMoves:
                s += i + ", "
        return s

    def __str__(self):
        return f"Name: {self.name}, HP: {self.hit_point}, weakness: {self.weakness}"

# code ends here

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())

print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
# Task 09

class Pokemon:

    def __init__(self, p):
        self.pokemon = p
        self.pokemon_type = "Needs to be set"
        self.pokemon_weakness = "Needs to be set"
    def kind(self):
        return self.pokemon_type     
    def weakness(self):
        return self.pokemon_weakness 
    def what_am_i(self):
        print("I am a Pokemon.")


# write your code here

class Pikachu(Pokemon):
    def __init__(self):
        super().__init__(p = "Pikachu")
        self.pokemon_type = "Electric"
        self.pokemon_weakness = "Ground"

    def what_am_i(self):
        super().what_am_i()
        print("I am a Pikachu")

class Charmander(Pokemon):
    def __init__(self):
        super().__init__(p = "Charmander")
        self.pokemon_type = "Fire"
        self.pokemon_weakness = "Water, Ground and Rock"

    def what_am_i(self):
        super().what_am_i()
        print("I am a Charmander")

# code ends here

pk1 = Pikachu()
print("Pokemon:", pk1.pokemon)
print("Type:", pk1.kind())
print("Weakness:", pk1.weakness())
pk1.what_am_i()
print("========================")
c1 = Charmander()
print("Pokemon:", c1.pokemon)
print("Type:", c1.kind())
print("Weakness:", c1.weakness())
c1.what_am_i()
```

```python
# Task 10
class Department:
    def __init__(self, s):
        self.semester = s
        self.name = "Default"
        self.id = -1
        
    def student_info(self):
        print("Name:", self.name)
        print("ID:", self.id) 
        
    def courses(self, c1, c2, c3):
        print("No courses Approved yet!")

# Write your code here
class CSE(Department):
    def __init__(self, sName, sId, sem):
        super().__init__(sem)
        self.name = sName
        self.id = sId

    def courses(self, c1, c2, c3):
        print(f"Courses Approved to this CSE student in {self.semester}:")
        print(f"{c1}\n{c2}\n{c3}")

class EEE(Department):
    def __init__(self, sName, sId, sem):
        super().__init__(sem)
        self.name = sName
        self.id = sId

    def courses(self, c1, c2, c3):
        print(f"Courses Approved to this EEE student in {self.semester}:")
        print(f"{c1}\n{c2}\n{c3}")
        

# code ends here

s1 = CSE("Rahim", 16101328,"Spring2016")
s1.student_info()
s1.courses("CSE110", "MAT110", "ENG101")
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.student_info()
s2.courses("Mat110", "PHY111", "ENG101")
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.student_info()
s3.courses("CSE111", "PHY101", "MAT120")
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.student_info()
s4.courses("EEE201", "PHY112", "MAT120")
```
