```python
# Task 03 : UB Jumper


while(True):
    x = input()
    if x == "STOP":
        break
    nums = list(map(int, x.split()))

    abs_value = []
    for i in range(len(nums)-1):
        abs_value.append(abs(nums[i]-nums[i+1]))

    is_ub_jumper = True
    for i in range(1, len(nums)):
        if i not in abs_value:
            is_ub_jumper = False
            break
    if is_ub_jumper:
        print("UB Jumper")
    else:
        print("Not UB Jumper")
    
```

## Dictionary & Tuple


```python
# Task 01: Combine two dictionary

d1 = input()
d2 = input()

# Creating dictionary from input
def makeDict(str_dict):
    str_dict = str_dict.split(", ")
 
    dict = {}
    for i in range(len(str_dict)):
        k, v = str_dict[i].split(": ")
        
        dict[k] = int(v)
        
    return dict

d1 = makeDict(d1)
d2 = makeDict(d2)

#===============================

for key, value in d1.items():

    if d2.get(key) != None:
        d1[key] += d2[key]

for key, value in d2.items():
    
    if d1.get(key) == None:
        d1[key] = d2[key]



unique = []
for value in d1.values():
    if value not in unique:
        unique.append(value)

# Sorting array named unique
for j in range(1, len(unique)):
    
    key = unique[j]
    i = j-1
    while(i>=0 and unique[i]>key):
        unique[i+1] = unique[i]   
        i = i-1
    unique[i+1] = key  

unique = tuple(unique)

print(d1)
print(f"Values: {unique}")
```

```python

# Task 01: BMI

def bmi(height, weight):

    res = (weight)/((height/100)**2)
    res = float("{:.1f}".format(res))
    # print("{:.1f}".format(res))

    message = {"18.5": "Underweight", "25": "Normal", "30": "Overweight"}


    isObese = True
    for key, value in message.items():

        if res < float(key):
            print(f"Score is {res}. You are {value}")
            isObese = False
            break
    if isObese == True:
        print(f"Score is {res}. You are Obese")
        


bmi(175, 96)
bmi(152, 48)
```

```python
# Task 2: Find divisors

def findDivisor(min, max, divisor):

    total = 0

    for i in range(min, max):
        if i%divisor ==0:
            total += i
    return total

ans = findDivisor(0,10,2)
print(ans)
print(findDivisor(3, 16, 3))
```

```python
# Task 03: Burger Cost

def cost(name, place = "Mohakhali"):

    menu = {"BBQ Chicken Cheese Burger": 250, "Beef Burger": 170, "Naga Drums":200}
    total = 0
    for burger, price in menu.items():
        # print(burger, price)
        if burger == name:
            if place != "Mohakhali":
                total = price + 60 + (price*8/100)
            else:
                total = price + 40 + (price*8/100)
    print(total)

cost("Beef Burger")
cost('Beef Burger', 'Dhanmondi')
```

```python
# Task 04: Update email with domain name

def replace_domain(email, new, old = "sheba.xyz"):

    username, ext = email.split("@")
    newEmail = username+ "@"+ new

    if ext == new and ext == old:
        print(f"Unchanged: {newEmail}")
    else:
        print(f"Changed: {newEmail}")
        

replace_domain('alice@kaaj.com', 'sheba.xyz', 'kaaj.com')
replace_domain('bob@sheba.xyz', 'sheba.xyz')
```

```python
# Task 05 : Palindrome checker

def isPalindrome(s):

    # removing whitespace (There are multiple ways)
    s = "".join(s.split(" "))
    flag = True
    if len(s)>1:
        j = len(s)-1
        
        for i in range(len(s)//2):
            if s[i] == s[j]:
                j-=1
                continue
            else:
                flag = False
                break
    if flag == True:
        print("Palindrome")
    else:
        print("Not a palindrome")

isPalindrome("nurses run")
```

```python
# Task 06: Capitalization

def capitalize(s):

    newS = ""
    for i in range(len(s)):
        if i == 0:
            newS += s[i].upper()

        elif (s[i] == 'i') and ((s[i-1] == " " and (s[i+1] == " " or s[i+1] == "." or s[i+1] == "?" or s[i+1]=="!")) or (s[i+1] == "'") ):

            newS += s[i].upper()
        
        elif (s[i-2] == '.' or s[i-2] == "?" or s[i-2] == "!"):
            newS += s[i].upper()

        else:
            newS += s[i]
        
        
    print(newS)
    
capitalize("should i've ? may i love my pet very. ")
capitalize("okay then i'll to do the task later. ")
capitalize("my name is rajo i? ")
```
