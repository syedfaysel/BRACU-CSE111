```python
# Task 02:

class Shape:
    def __init__(self, shape, param1 , param2):
        self.shape = shape
        self.a = param1
        self.b = param2
        
    
    def area(self):
        ans = "Shape unknown"
        if self.shape == "Triangle" or self.shape == "Rhombus":
            ans = self.a*self.b*0.5
        elif self.shape == "Square" or self.shape == "Rectangle":
            ans = self.a * self.b

        print(f"Area: {ans}")
            

#=================================

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()

```

```python
# Task 03

class Calculator:
    def __init__(self):
        print("Calculator is ready!")

    def calculate(self, n1, n2, sign):
        self.n1 = n1
        self.n2 = n2
        self.sign = sign
        self.ans = None
        if sign == '+':
            self.ans = n1+n2
        elif sign == '-':
            self.ans = n1 - n2
        elif sign == '*':
            self.ans = n1 * n2
        elif sign == '/':
            self.ans = n1/n2

        return self.ans

    def showCalculation(self):
        print(f"{self.n1} {self.sign} {self.n2} = {self.ans}")


#=============================================
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
# Task 09
class Batsman:
    def __init__(self, *args):
        if len(args) == 3:
            self.name = args[0]
            self.run = args[1]
            self.ball = args[2]
        else:
            self.name = "New Batsman"
            self.run = args[0]
            self.ball = args[1]


    def printCareerStatistics(self):
        print(f"Name: {self.name}")
        print(f"Runs Scored: {self.run} , Balls Faced: {self.ball}")
    
    def battingStrikeRate(self):
        # Batting strike rate (s/r) = runsScored / ballsFaced x 100.
        return self.run / self.ball * 100

    def setName(self, name):
        self.name = name

# =========== Driver Code ============

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
# Task 12 

class Account:
    def __init__(self, name = None, balance = 0.0):
        self.name = name
        self.balance = float(balance)

    def details(self):
        if self.name == None:
            return f"Default Account\n{self.balance}"
        else:
            return f"{self.name}\n{self.balance}"  


    def withdraw(self, amount):
        if (self.balance-amount <=3070):
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        else:
            self.balance -= amount
            print(f'Withdraw successful! New balance is: {self.balance}')
#=========================

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
# Task 13
# Design code
class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}

    def calculateGPA(self, gradeSheet, sem):
        courses = []
        gp_list = []
        for item in gradeSheet:
            course, gp = item.split(": ")
            gp = float(gp)
            courses.append(course)
            gp_list.append(gp)
        
        # GPA = Sum of (Grade Points * Credits)/ Credits attempted
        gpa = sum(gp_list) / len(gp_list)
        gpa = ("{:.2f}".format(gpa))

        courses = tuple(courses)
        self.grades[sem] = {courses:gpa}

    def printDetails(self):
        print("Name: ", self.name)
        print("ID: ", self.id)
        
        for sem, res in self.grades.items():
            print(f"Courses taken in {sem}:")

            for courses in self.grades[sem]:
                for course in courses:
                    print(course)
                print(self.grades[sem][courses])
            

# Driver code

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python

# Task 25 : So called difficult proble (Actually it is)

class Vaccine:
    def __init__(self, name, origin, day):
        self.name = name
        self.origin = origin
        self.day = day


class Person:
    def __init__(self, name, age, prof = "General Citizen"):
        self.name = name
        self.age = age
        self.prof = prof

        self.vaccineDose = []

        self.data = {1:{"status": 0}, 2: {"status": 0}}



    def pushVaccine(self, vacObject, dose = None):

        flag = True
        if self.age < 25 and self.prof != "Student":
            flag = False
            print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now.")
        
        if dose == None and self.data[1]["status"] == 0 and flag:
            self.data[1]["status"] = 1
            self.vaccinated = vacObject
            print(f"1st dose done for {self.name}")
        else:
            if self.data[1]['status'] == 1 and self.data[2]["status"] == 0 and flag:
                if self.vaccinated.name != vacObject.name:
                    print(f"Sorry {self.name}, you can’t take 2 different vaccines")
                else:
                    self.data[2]["status"] = 1
                    print(f"2nd dose done for {self.name}")


    def showDetail(self):
        
        print(f"Name: {self.name} Age: {self.age} Type: {self.prof}")
        print(f"Vaccine name: {self.vaccinated.name}")
        if self.data[1]['status'] == 1:
            print(f"1st dose: Given")
        if self.data[1]['status'] == 1 and self.data[2]['status'] == 0:
            print(f"2nd dose: Please come after {self.vaccinated.day} days")
        else:
            print("2nd dose: Given")



astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```

```python
# Solve for quiz 02

class ClonePc:
    def __init__(self, pctype ="Default"):
        self.pctype = pctype

        self.features = {}

    def addFeature(self, key, value):
        
        if self.pctype == "Default":
            print("Features cannot be added with no PCtype")
        else:
            if key in self.features:
                self.features[key].append(value)
            else:
                self.features[key] = [value]

    def setType(self, pctype):
        self.pctype = pctype

    def giveDetail(self):
        message = ""
        for k, v in self.features.items():
            message += f"{k} :"
            for i in v:
                message += f"{i},"
            message +='\n'
        return message



    def printDetail(self):
        detail = self.giveDetail()
        print(detail)


# ================ Driver Code =================
c1 = ClonePc()
c1.addFeature("CPU", "Intel i7")

print("========================")

c1.setType("Windwos Pc")
c1.addFeature("CPU", "Intel i7")
c1.addFeature("RAM", "DDR4")
c1.addFeature("RAM", "12 GB")
c1.addFeature("Monitor", "21 inch")

print(c1.giveDetail())
c1.printDetail()
```
